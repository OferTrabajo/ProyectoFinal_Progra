# Notebook 1: Exploración de datos

Este notebook realiza el análisis exploratorio del dataset de transacciones con tarjeta de crédito para detectar fraudes.

Pasos:
1. Carga del dataset
2. Exploración inicial (shape, head, estadísticos descriptivos)
3. Distribución de la variable `Amount`
4. Distribución de la clase objetivo `Class`

6. Detección de outliers con boxplot


## 1. Carga del dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Carga del dataset, en este caso es de Transacciones anónimas con tarjeta de crédito etiquetadas como fraudulentas o genuinas
# Sacado de https://www.kaggle.com/datasets/mlg-ulb/creditcardfraud

df = pd.read_csv('creditcard.csv')
print('Shape del dataset:', df.shape)
df.head()
```

## 2. Exploración inicial

Mostramos información básica para entender la estructura y contenido de los datos.


```python
# Información general del DataFrame
df.info()

# Estadisticas descriptivos de variables unica y exclusivamente numericas
df.describe()
```

## 3. Distribución de la variable `Amount`

Visualizamos cómo se distribuyen los montos de las transacciones.


```python
plt.figure(figsize=(8, 5))
sns.histplot(df['Amount'], bins=50, kde=True)
plt.title('Distribución de Amount')
plt.xlabel('Amount')
plt.ylabel('Frecuencia')
plt.show()
```

## 4. Distribución de la clase objetivo `Class`

Contamos cuántas transacciones fraudulentas vs. no fraudulentas existen y mostramos en gráfico.


```python
# Conteo de cada clase
conteo_clases = df['Class'].value_counts()
print(conteo_clases)

# Gráfico de barras
plt.figure(figsize=(6, 4))
sns.barplot(x=conteo_clases.index, y=conteo_clases.values)
plt.xticks([0,1], ['No fraude (0)', 'Fraude (1)'])
plt.ylabel('Cantidad de transacciones')
plt.title('Distribución de clases')
plt.show()
```

## 5. Mapa de calor de correlaciones

Identificamos correlaciones entre las variables y la clase objetivo.


```python
# Calculamos matriz de correlaciones
corr = df.corr()

# Filtramos correlaciones con respecto a 'Class' ordenadas de mayor a menor
corr_target = corr['Class'].sort_values(ascending=False)
print('Correlaciones con Class:')
print(corr_target)

# Mapa de calor completo (puede ser pesado, pero útil para ver patrones generales)
plt.figure(figsize=(16, 12))
sns.heatmap(corr, cmap='coolwarm', linewidths=0.5)
plt.title('Mapa de calor de correlaciones (todas las variables)')
plt.show()
```

## 6. Detección de outliers con boxplot

Revisamos posibles valores atípicos en la variable `Amount`.


```python
plt.figure(figsize=(6, 4))
sns.boxplot(x=df['Amount'])
plt.title('Boxplot de Amount')
plt.show()
```

## 7. Análisis de transacciones por hora (opcional)

Si existe la columna `Time`, convertimos a hora del día y contamos transacciones por hora para ver si hay picos de fraude en ciertos momentos.


```python
if 'Time' in df.columns:
    # Convertimos segundos desde la primera transacción a hora del día
    df['Hour'] = (df['Time'] // 3600) % 24

    # Conteo de transacciones totales por hora
    conteo_horas = df.groupby('Hour').size()
    conteo_fraude_horas = df[df['Class'] == 1].groupby('Hour').size()

    plt.figure(figsize=(10, 4))
    conteo_horas.plot(kind='line', marker='o', label='Todas las transacciones')
    conteo_fraude_horas.plot(kind='line', marker='o', label='Transacciones fraudulentas')
    plt.xlabel('Hora del día')
    plt.ylabel('Cantidad de transacciones')
    plt.title('Transacciones por hora')
    plt.legend()
    plt.show()
else:
    print('La columna Time no está presente en el dataset.')
```
