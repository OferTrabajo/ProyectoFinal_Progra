```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar el dataset, en este caso seleccione para el proyecto: Transacciones anónimas con tarjeta de crédito etiquetadas como fraudulentas o genuinas.
# Sacado de: https://www.kaggle.com/datasets/mlg-ulb/creditcardfraud

df = pd.read_csv('creditcard.csv')

# Vista general
print("Dimensiones del dataset:", df.shape)
df.head()
```

```python
# Resumen estadistico.
df.describe()
```

```python
# Ver distribución de clases
print(df['Class'].value_counts())

sns.countplot(data=df, x='Class')
plt.title('Distribución de clases (0 = No Fraude, 1 = Fraude)')
plt.show()
```

```python
# Verificamos valores nulos, como en este caso no hay ninguno, los simularemos aleatoriamente el el preprocesado.
print("Valores nulos por columna:\n", df.isnull().sum())
```

```python
# Histogramas de algunas variables
import matplotlib.pyplot as plt

columnas_a_graficar = ['V1', 'V2', 'V3', 'Amount']
df[columnas_a_graficar].hist(figsize=(10,8), bins=50)
plt.tight_layout()
plt.show()
```

```python
# Mapa de calor de correlaciones con la variable Class
correlaciones = df.corr()
plt.figure(figsize=(12, 8))
sns.heatmap(correlaciones[['Class']].sort_values(by='Class', ascending=False), annot=True, cmap='coolwarm')
plt.title("Correlaciones con la variable Class")
plt.show()
```
